```python
def get_inf_dt(url, content, len_of_page):
  try:
    df=[]
    for cont in lst_content:
      for i in range(len_of_page):
        url_ = url + cont.split('.')[0] + f'/trang-{i+1}.htm'
        print(url_)
        page = requests.get(url_)
        soup = BeautifulSoup(page.content, 'html.parser')
        new_feed = soup.findAll('a', class_='dt-text-black-mine')
        print(new_feed)
        for feed in new_feed:
          dic={}
          href = feed.get('href')
          new_url = url + href
          print(new_url)
          page = requests.get(new_url)
          soup = BeautifulSoup(page.content, 'html.parser')

          req_title = soup.find('h1', class_='title-page detail')
          if req_title is None:
            break
          else:
            title = req_title.getText()

          x = soup.find(class_='singular-sapo')
          if x is None:
            break
          else:
            x = x.getText()
            x = x.replace('(Dân trí) - ','')
          req_y = soup.findAll('p')
          y=''
          for req in req_y:
            y+=str(req.getText())
          ### dong goi
          dic['title'] = title
          dic['document'] = y
          dic['summary'] = x
          df.append(dic)
        print(f'-----------done page {i+1}------------')
      print(f'--------------------content {cont}-------------------------')
      if df is None:
        print('Error.........')
      else:
        print('continue......', len(df))
  except:
    pass
  return df
```

```python
url = 'https://dantri.com.vn'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
new_feed = soup.findAll('ol', class_='submenu')

lst_content = []
for feed in new_feed:
  for ele in feed.findAll('a'):
    if 'https' not in ele.get('href'):
      lst_content.append(ele.get('href'))

# print(lst_content)
df = get_inf_dt(url, lst_content, 3)
```
